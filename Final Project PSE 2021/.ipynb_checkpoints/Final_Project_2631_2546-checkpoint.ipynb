## Merge the two datasets

## Cleaning Text 

```python
# return the wordnet object value corresponding to the POS tag
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('vader_lexicon')
nltk.download('averaged_perceptron_tagger')
from nltk.corpus import wordnet

def get_wordnet_pos(pos_tag):
    if pos_tag.startswith('J'):
        return wordnet.ADJ
    elif pos_tag.startswith('V'):
        return wordnet.VERB
    elif pos_tag.startswith('N'):
        return wordnet.NOUN
    elif pos_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
    
import string

from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.tokenize import WhitespaceTokenizer
from nltk.stem import WordNetLemmatizer

def clean_text(text):
    # lower text
    text = text.lower()
    # tokenize text and remove puncutation
    text = [word.strip(string.punctuation) for word in text.split(" ")]
    # remove words that contain numbers
    text = [word for word in text if not any(c.isdigit() for c in word)]
    # remove stop words
    stop = stopwords.words('english')
    text = [x for x in text if x not in stop]
    # remove empty tokens
    text = [t for t in text if len(t) > 0]
    # pos tag text
    pos_tags = pos_tag(text)
    # lemmatize text
    text = [WordNetLemmatizer().lemmatize(t[0], get_wordnet_pos(t[1])) for t in pos_tags]
    # remove words with only one letter
    text = [t for t in text if len(t) > 1]
    # join all
    text = " ".join(text)
    return(text)

# clean text data
full_reviews["Description_clean"] = full_reviews["Description"].apply(lambda x: clean_text(x))
full_reviews.head(10)
```

## Vader 

```python
# add sentiment anaylsis columns
from nltk.sentiment.vader import SentimentIntensityAnalyzer

full_reviews_vader=full_reviews.copy()
for index, row in full_reviews_vader['Description_clean'].iteritems():
    score = SentimentIntensityAnalyzer().polarity_scores(row)
    neg = score['neg']
    neu = score['neu']
    pos = score['pos']
    comp = score['compound']
    if neg > pos:
        full_reviews_vader.loc[index, 'sentiment'] = "negative"
    elif pos > neg:
        full_reviews_vader.loc[index, 'sentiment'] = "positive"
    else:
        full_reviews_vader.loc[index, 'sentiment'] = "neutral"
    full_reviews_vader.loc[index, 'neg'] = neg
    full_reviews_vader.loc[index, 'neu'] = neu
    full_reviews_vader.loc[index, 'pos'] = pos
    full_reviews_vader.loc[index, 'compound'] = comp

full_reviews_vader.head(10)
```

```python
def count_values_in_column(data,feature):
    total=data.loc[:,feature].value_counts(dropna=False)
    percentage=round(data.loc[:,feature].value_counts(dropna=False,normalize=True)*100,2)
    return pd.concat([total,percentage],axis=1,keys=['Total','Percentage'])

#Count_values for sentiment
count_values_in_column(full_reviews_vader,"sentiment")
```

```python
pichart = count_values_in_column(full_reviews_vader,"sentiment")
labels = [  'Neutral ['+str(pichart["Percentage"]["neutral"])+'%]','Positive ['+str(pichart["Percentage"]["positive"])+'%]','Negative ['+str(pichart["Percentage"]["negative"])+'%]']
sizes=[]
sizes.append(pichart["Percentage"]["neutral"])
sizes.append(pichart["Percentage"]["positive"])
sizes.append(pichart["Percentage"]["negative"])
colors = [ 'blue','green','red']
patches, texts = plt.pie(sizes,colors=colors, startangle=90)
plt.style.use('default')
plt.legend(labels)
plt.title("Sentiment Analysis Result for  Restaurants and Hotels" )
plt.axis('equal')
plt.show()
```

## TF-IDF 

```python
tfidf = TfidfVectorizer(analyzer='word', ngram_range=(1, 3), min_df=1)
X = tfidf.fit_transform(full_reviews["Description_clean"])
```

## Split and Shape of train and test datasets

```python
X_train,X_test,y_train,y_test=train_test_split(X,full_reviews["Rating_Bin"],test_size=0.2,random_state=42)
print(X_train.shape)
print(X_test.shape)
```

## Machine Learning Algorithms

## Initialze the estimators for GridSearchCV

```python
# Initialze the estimators
clf1 = RandomForestClassifier(random_state=42)
clf2 = SVC(probability=True, random_state=42)
clf3 = LogisticRegression(random_state=42)
clf4 = XGBClassifier(max_depth=10)
clf5 = KNeighborsClassifier()
clf6 = MultinomialNB()
clf7=autosklearn.classification.AutoSklearnClassifier(time_left_for_this_task=2*60, per_run_time_limit=30, n_jobs=8)



clf8 = Sequential()

# Adding the first LSTM layer and some Dropout regularisation
clf8.add(LSTM(units = 50,return_sequences = True,input_shape = (X_train.shape[1], 1)))
clf8.add(Dropout(0.1))

# Adding a second LSTM layer and some Dropout regularisation
clf8.add(LSTM(units = 50, return_sequences = True))
clf8.add(Dropout(0.1))

# Adding a fourth LSTM layer and some Dropout regularisation
clf8.add(LSTM(units = 50))
clf8.add(Dropout(0.1))

# Adding the output layer
clf8.add(Dense(units = 1))

optimizer = optimizers.Adam(learning_rate=0.001 , beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)

clf8.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics=['accuracy'])

clf8 = KerasClassifier(clf8)
```

## Setting parameters for each model

```python
# Initiaze the hyperparameters for each dictionary
param1 = {}
param1['classifier__n_estimators'] = [10, 50, 100, 250]
param1['classifier__criterion'] = ["gini", "entropy"]
param1['classifier'] = [clf1]

param2 = {}
param2['classifier__C'] = [10**-1, 10**0, 10**1, 10**2]
param2['classifier__kernel'] = ["linear","rbf"]
param2['classifier'] = [clf2]

param3 = {}
param3['classifier__C'] = [10**-1, 10**0, 10**1, 10**2]
param3['classifier'] = [clf3]

param4 = {}
param4['classifier'] = [clf4]

param5 = {}
param5['classifier__n_neighbors'] = [2,5,10,25,50]
param5['classifier'] = [clf5]

param6 = {}
param6['classifier__alpha'] = [10**-3,10**-2, 10**-1, 1]
param6['classifier'] = [clf6]

param7 = {}
param7['classifier'] = [clf7]

param8 = {}
param8['classifier'] = [clf8]

```

## Train the grid search model

```python
pipeline = Pipeline([('classifier', clf1)])
params = [param1, param2, param3,param4, param5,param6,param7,param8]
# Train the grid search model
gs = GridSearchCV(pipeline,params, cv=5, scoring='roc_auc').fit(X_train, y_train)
```

```python
X_train,X_test,y_train,y_test=train_test_split(full_reviews["Description"],full_reviews["Rating_Bin"],test_size=0.2,random_state=42)
```

## Bert 

```python
bert_preprocess = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3")
bert_encoder = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4")
```

## Configuration of the Model

```python
# Bert layers
text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')
preprocessed_text = bert_preprocess(text_input)
outputs = bert_encoder(preprocessed_text)

# Neural network layers
l = tf.keras.layers.Dropout(0.1, name="dropout")(outputs['pooled_output'])
l = tf.keras.layers.Dense(1, activation='sigmoid', name="output")(l)

# Use inputs and outputs to construct a final model
model = tf.keras.Model(inputs=[text_input], outputs = [l])

model.summary()
```

## Metrics for evaluation

```python
METRICS = [
      tf.keras.metrics.BinaryAccuracy(name='accuracy'),
      tf.keras.metrics.Precision(name='precision'),
      tf.keras.metrics.Recall(name='recall')
]

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=METRICS)
```

## Training of the Model

```python
model.fit(X_train, y_train, epochs=2)
```

## Evaluation of the Model

```python
model.evaluate(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)
y_predicted = y_predicted.flatten()

y_predicted = np.where(y_predicted > 0.5, 1, 0)
```
