```python
import pandas as pd

result = pd.read_excel("./sales.xlsx")

result.head()
```

## Removing the rows with both nan values of Quant and Val

```python

#result.dropna(thresh=2,subset=['Quant', 'Val'])
result=result.dropna()
print(result.size)
```

## Removing the outliers with the use of the IQR method

```python
Q1 = result.quantile(0.25)
Q3 = result.quantile(0.75)
IQR = Q3 - Q1

result = result[~((result < (Q1 - 1.5 * IQR)) |(result > (Q3 + 1.5 * IQR))).any(axis=1)]
result.size
```

## Clustering the data

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
  
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
result["Insp_code"] = ord_enc.fit_transform(result[["Insp"]])

ord_enc = OrdinalEncoder()
result["Prod_code"] = ord_enc.fit_transform(result[["Prod"]])

result["Unit_price"]=result[["Val"]].div(result.Quant, axis=0)
#for i in range (result["Unit_price"].size):
    #np.isinf(result.at[i,"Unit_price"])|
 #   result["Unit_price"]=result[result["Unit_price"].values!="unkn"]


kmeans = KMeans(n_clusters=3).fit(result[["Unit_price","Prod_code"]])
centroids = kmeans.cluster_centers_
print(centroids)

plt.scatter(result['Unit_price'],result['Prod'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```

```python
from sklearn.decomposition import PCA

ord_enc = OrdinalEncoder()
result["Prod_code"] = ord_enc.fit_transform(result[["Prod"]])
print(result[["Prod", "Prod_code"]].head(11))

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(result[["Prod_code","Quant","Val"]])
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['PCA1', 'PCA2'])

finalDf = pd.concat([principalDf, result[['Insp']]], axis = 1)


fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('PCA1', fontsize = 15)
ax.set_ylabel('PCA2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['unkn', 'fraud', 'ok']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['Insp'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'PCA1']
               , finalDf.loc[indicesToKeep, 'PCA2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

## Semi Supervised learning Model

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn import svm




Labeled_data=result[result["Insp"].values!="unkn"]

Labeled_data
```

```python
X_train, X_test, y_train, y_test = train_test_split( Labeled_data[["Prod_code","Quant","Val"]],
                                                    Labeled_data[["Insp_code"]], test_size=0.30, random_state=42)

# all parameters not specified are set to their defaults
print(y_train.shape)
print(X_train.shape)

#clf = svm.SVC(kernel='linear', probability=True,C=1).fit(X_train, y_train)
print(y_train.shape)
```

```python
clf = svm.SVC(kernel='linear', probability=True,C=1).fit(X_train, y_train.values.ravel())
```

```python
clf.score(X_test, y_test)
```

```python
Unlabeled_data=result[result["Insp"].values=="unkn"]
X_unlabeled=Unlabeled_data[["Prod_code","Quant","Val"]]

clp= clf.predict_proba(X_unlabeled)
pseudo_label=clf.predict(X_unlabeled)

#Unlabeled_data = pd.DataFrame(clp, columns = ['C1Prob', 'C2Prob']) 
Unlabeled_data['C1Prob'] = clp.T[0]
Unlabeled_data['C2Prob'] = clp.T[1] 
Unlabeled_data['Insp_code']=pseudo_label
Unlabeled_data['max']=Unlabeled_data[["C1Prob", "C2Prob"]].max(axis=1)
Unlabeled_data
```

```python
frames = [Unlabeled_data[["Prod_code","Quant","Val","Insp_code"]], Labeled_data[["Prod_code","Quant","Val","Insp_code"]]]

final_data = pd.concat(frames)
final_data
```

```python
X_train, X_test, y_train, y_test = train_test_split( final_data[["Prod_code","Quant","Val"]],
                                                    final_data[["Insp_code"]], test_size=0.30, random_state=42)

clf = svm.SVC(kernel='linear', probability=True,C=1).fit(X_train, y_train.values.ravel())

clf.score(X_test, y_test)
```
